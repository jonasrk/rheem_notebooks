```python
%matplotlib inline

import matplotlib.pyplot
import pylab

LABEL_COLOR_MAP = {0 : 'r',
                   1 : 'k',
                   2 : 'b',
                   3 : 'g',
                   4 : 'c',}

file_object  = open('/Users/jonas/tmp_kmeans.txt', 'r') 

centroids_x = 0.1564, 0.8333, 0.7545, 0.2512, 0.4726
centroids_y = 0.2655, 0.2765, 0.7785, 0.7653, 0.2337

x = []
y = []
color = []

for line in file_object: 
    x.append(float(line.strip().split(', ')[0]))
    y.append(float(line.strip().split(', ')[1]))
    
for i in range(0, len(x)):
    min_distance = 2
    nearest_centroid = -1
    for j in range(0, len(centroids_x)):
        distance = ((x[i] - centroids_x[j])**2 + (y[i] - centroids_y[j])**2)**0.5
        if (distance < min_distance):
            min_distance = distance
            nearest_centroid = j
    color.append(LABEL_COLOR_MAP[nearest_centroid])
    
matplotlib.pyplot.scatter(x,y, c=color)
matplotlib.pyplot.show()
```
